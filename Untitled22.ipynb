```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#set index:
df.set_index('#').head()
```

```python
#countries with biggest biggest consumption.
df['Daily Oil Consumption (Barrels)'].max()
```

```python
#top 10 countries with the highest prices per litre.
df[['Country','Price Per Liter (USD)']].nlargest(n=10,columns='Price Per Liter (USD)').plot(kind='bar',color=sns.color_palette('magma'))
```

```python
#top 10 countries with the least prices per litre.
df[['Country','Price Per Liter (USD)']].nsmallest(n=10,columns='Price Per Liter (USD)')
```
